```python
from __future__ import division
import nltk, re, pprint
```

```python
raw_chat_file_obj = open("DotaChat.txt", 'r', encoding='utf-8')
```

```python
raw_chat_text = raw_chat_file_obj.readlines()
```

```python
type(raw_chat_text)
```

```python
chat_str = ""

for line in raw_chat_text:
    chat_str += line.replace("\n", " ")
```

```python
len(chat_str)
```

```python
type(chat_str)
```

```python
tokens = nltk.word_tokenize(chat_str)
```

```python
tokens[:50]
```

```python
nltk_text = nltk.Text(tokens)
```

```python
type(nltk_text)
```

```python
nltk_text[1020:1040]
```

```python
'; '.join(nltk_text.collocation_list()) # need this, otherwise error, too many values to unpack.
```

```python
nltk_text.concordance('dota')
```

```python
nltk_text.generate()
```

```python
text = nltk_text
```

```python
text.similar("dota")
```

```python
text.common_contexts(["saqibur", "akib"])
```

```python
text.dispersion_plot(["saqibur", "akib", "sharaf", "kinjol", "rubab"])
%matplotlib inline
```

```python
fdist = nltk.FreqDist(text)
```
